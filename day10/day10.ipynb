# Day 10: Adapter Array

[Brief](https://adventofcode.com/2020/day/10)

```python
def get_differences(l, current=0):
    diffs = []
    end = max(l) + 3
    
    for i in range(1, 4):        
        if current + i == end:
            return [3]
        elif current + i > end:
            return None
        
        if current + i not in l:
            continue
        
        diffs = get_differences(l, current=current + i)
        
        if diffs is None:
            continue
        
        diffs.append(i)
        return diffs
    
    return None
```

## Examples

```python
with open("example.txt", "r") as file:
    example_joltages = [int(j) for j in file.read().splitlines()]

example_diffs = get_differences(example_joltages)
assert example_diffs.count(1) == 7
assert example_diffs.count(3) == 5
```

```python
with open("example2.txt", "r") as file:
    example2_joltages = [int(j) for j in file.read().splitlines()]

example2_diffs = get_differences(example2_joltages)
assert example2_diffs.count(1) == 22
assert example2_diffs.count(3) == 10
```

## Part 1

```python
with open("input.txt", "r") as file:
    joltages = [int(j) for j in file.read().splitlines()]

diffs = get_differences(joltages)
print(diffs.count(1) * diffs.count(3))
```

## Part 2

```python
def get_combinations(l, current=0):
    total = 0
    end = max(l) + 3
    
    for i in range(1, 4):        
        if current + i == end:
            total += 1
        elif current + i > end:
            continue
        
        if current + i not in l:
            continue
        
        total += get_combinations(l, current=current + i)
        
    return total
```

```python
assert get_combinations(example_joltages) == 8
assert get_combinations(example2_joltages) == 19208
```
